```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process

def fuzzy_match_and_merge(candidate_data, ceda_data):
    # Function to find the best fuzzy match for each candidate in the CEDA dataset
    def find_best_match(name, candidates):
        match, score,_ = process.extractOne(name, candidates, scorer=fuzz.token_set_ratio)
        return match, score

    # Create a dictionary of candidate names as keys and their priorities as values
    name_to_priority = {}
    name_to_score = {}

    num_added = 0
    num_read = 0

    for _, row in ceda_data.iterrows():
        ceda_name = f"{row['first']} {row['last']}"
        ceda_year = row['year']
        ceda_county = row['cntyname']

        # Filter candidate data to the same year and county
        year_county_matched_candidates = candidate_data[(candidate_data['year'] == ceda_year) & (candidate_data['county'] == ceda_county)]

        if not year_county_matched_candidates.empty:
            match, score = find_best_match(ceda_name, year_county_matched_candidates['candidate_name'])
            if score > 80:
                # Found a match with a high enough score, add to dictionary
                name_to_priority[ceda_name] = year_county_matched_candidates.loc[
                    year_county_matched_candidates['candidate_name'] == match, 'priority'].values[0]
                name_to_score[ceda_name] = score  # Store the score
                num_added += 1

                if score != 100:
                  print("Name in CEDA data: ", ceda_name)
                  print("Matched name from candidate data: ", match)
                  print("Year: ", ceda_year)
                  print("Score was: ", score)
        num_read += 1

        if num_read % 100 == 0:
          print(num_added , "samples have been matched out of ", num_read)


    # Create a new 'priorities' column in the CEDA dataset
    ceda_data['priorities'] = ceda_data.apply(
        lambda row: name_to_priority.get(f"{row['first']} {row['last']}", None), axis=1
    )

    # Create a new 'score' column in the CEDA dataset
    ceda_data['score'] = ceda_data.apply(
        lambda row: name_to_score.get(f"{row['first']} {row['last']}", None), axis=1
    )


    return ceda_data

# Call the fuzzy_match_and_merge function to merge the datasets
#copy = CEDA.head(1000).copy()
merged_ceda_data = fuzzy_match_and_merge(candidate_data, CEDA)
```

```python

# Load the data
file_path = '/content/drive/My Drive/Priorities_Latest_10-10-23.csv'
df = pd.read_csv(file_path)
df = df[~df['priority'].str.startswith("NA")]
docs = df['priority'].tolist()

# Cleaning and preprocessing
stop = set(stopwords.words('english'))
exclude = set(string.punctuation)
lemma = WordNetLemmatizer()

def clean(doc):
    stop_free = " ".join([i for i in doc.lower().split() if i not in stop])
    punc_free = ''.join(ch for ch in stop_free if ch not in exclude)
    normalized = " ".join(lemma.lemmatize(word) for word in punc_free.split())
    return normalized

doc_clean = [clean(doc).split() for doc in docs]

# Preparing Document-Term Matrix
dictionary = corpora.Dictionary(doc_clean)
doc_term_matrix = [dictionary.doc2bow(doc) for doc in doc_clean]

# Initialize LDA Model
lda = LdaModel(corpus=doc_term_matrix, id2word=dictionary, num_topics=5, random_state=100, update_every=1, alpha='auto', eta='auto', passes = 40)

# # Initialize variables for coherence score plotting
# coherence_values = []
# pass_count = 10
# max_passes = 100

# for passes in range(10, max_passes + 1, pass_count):
#     lda.update(doc_term_matrix, passes=pass_count)
#     coherence_model_lda = CoherenceModel(model=lda, texts=doc_clean, dictionary=dictionary, coherence='c_v')
#     coherence_lda = coherence_model_lda.get_coherence()
#     coherence_values.append(coherence_lda)
#     print(f"Passes: {passes}, Coherence: {coherence_lda}")

# # Plotting coherence
# plt.plot(range(10, max_passes + 1, pass_count), coherence_values)
# plt.xlabel("Passes")
# plt.ylabel("Coherence score")
# plt.title("LDA Model Coherence")
# plt.show()

# Assign topics to documents
def assign_topics(ldamodel, corpus):
    topic_probs = []
    for doc in corpus:
        topic_prob = ldamodel.get_document_topics(doc, minimum_probability=0)
        # Ensure all topics are represented
        topic_dict = {topic: prob for topic, prob in topic_prob}
        full_topic_prob = [topic_dict.get(i, 0) for i in range(5)]
        topic_probs.append(full_topic_prob)
    return topic_probs

doc_topics = assign_topics(lda, doc_term_matrix)

# Create a DataFrame for topic probabilities
topic_df = pd.DataFrame(doc_topics, columns=[f'Topic_{i}' for i in range(5)])

# Concatenate with the original DataFrame
df = pd.concat([df, topic_df], axis=1)

# Save the updated DataFrame to a new CSV file
df.to_csv('/content/drive/My Drive/Priorities_Latest_10-10-23_with_Five_Topics.csv', index=False)

# Display the first few rows of the updated DataFrame
print(df.head())


# Visualize the topics
try:
    vis = gensimvis.prepare(lda, doc_term_matrix, dictionary)
    pyLDAvis.display(vis)
except Exception as e:
    print("An error occurred during LDA visualization:", e)
```

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
import string
from gensim import corpora
from gensim.models import LdaModel
from gensim.models.coherencemodel import CoherenceModel
import matplotlib.pyplot as plt
import pyLDAvis
import pyLDAvis.gensim_models as gensimvis
```

```python

# Download necessary NLTK data
nltk.download('stopwords')
nltk.download('wordnet')

# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Load the data
file_path = '/content/drive/My Drive/Priorities_Latest_10-10-23.csv'
df = pd.read_csv(file_path)
df = df[~df['priority'].str.startswith("NA")]
docs = df['priority'].tolist()

# Cleaning and preprocessing
stop = set(stopwords.words('english'))
exclude = set(string.punctuation)
lemma = WordNetLemmatizer()

def clean(doc):
    stop_free = " ".join([i for i in doc.lower().split() if i not in stop])
    punc_free = ''.join(ch for ch in stop_free if ch not in exclude)
    normalized = " ".join(lemma.lemmatize(word) for word in punc_free.split())
    return normalized

doc_clean = [clean(doc).split() for doc in docs]

# Preparing Document-Term Matrix
dictionary = corpora.Dictionary(doc_clean)
doc_term_matrix = [dictionary.doc2bow(doc) for doc in doc_clean]

# Grid search over the number of topics
num_topics_list = [3, 5, 8, 10, 13, 15, 17]
coherence_scores = []
perplexity_scores = []

for num_topics in num_topics_list:
    lda = LdaModel(doc_term_matrix, num_topics=num_topics, id2word=dictionary, passes=40, random_state=100)
    coherence_model_lda = CoherenceModel(model=lda, texts=doc_clean, dictionary=dictionary, coherence='u_mass')
    coherence_lda = coherence_model_lda.get_coherence()
    coherence_scores.append(coherence_lda)

    perplexity_lda = lda.log_perplexity(doc_term_matrix)
    perplexity_scores.append(perplexity_lda)

    print(f"Num Topics: {num_topics}, Coherence: {coherence_lda}, Perplexity: {perplexity_lda}")

# Plotting coherence and perplexity scores
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Number of Topics')
ax1.set_ylabel('Coherence Score', color=color)
ax1.plot(num_topics_list, coherence_scores, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:blue'
ax2.set_ylabel('Perplexity Score', color=color)  # we already handled the x-label with ax1
ax2.plot(num_topics_list, perplexity_scores, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title("LDA Model Coherence and Perplexity for Different Number of Topics")
plt.show()

# Choose the model with the highest coherence
best_num_topics = num_topics_list[perplexity_scores.index(max(perplexity_scores))]
print(f"Best Number of Topics: {best_num_topics}")

# Train the best model
best_lda_model = LdaModel(doc_term_matrix, num_topics=best_num_topics, id2word=dictionary, passes=400, random_state=100)

# Visualize the best topics
best_vis = gensimvis.prepare(best_lda_model, doc_term_matrix, dictionary)
pyLDAvis.display(best_vis)
```
